```python
#Subsoal 1

len(df.loc[(df["Attrition_Flag"] == "Attrited Customer") & (df["Customer_Age"] > 50)]) #Banyaknya attrited customer > 50 tahun
#dicetak menggunakan len()
```

```python
#Subsoal 2
nilai_kred_ter = df["Credit_Limit"].max() #Menyimpan nominal kartu kredit terbesar ke dalam variabel

print(df.loc[(df["Credit_Limit"] == nilai_kred_ter)]) #Memfilter dataframe yang ada
#terdapat lebih dari 1 orang yang memiliki nilai Credit_Limit terbesar
```

```python
#Subsoal 3

len(df.loc[(df["Card_Category"] == "Blue") & (df["Months_on_book"] > 40)]) #Banyaknya kartu dengan kategori Blue serta Months_on_book > 40 bulan
#dicetak menggunakan len()
```

```python
#Subsoal 4
df4 = df.loc[df["Total_Revolving_Bal"] == 0] #Memfilter data dengan Total_Revolving_Bal = 0
df4["Income_Category"].value_counts() #Mencetak jumlahnya berdasarkan kategori
```

```python
#Subsoal 5
df5 = df.loc[df["Attrition_Flag"] == "Existing Customer"] #Menyimpan data yang hanya Attrition_Flag == Existing Customer ke dalam variabel
df5.sort_values(["Months_Inactive_12_mon"], ascending = [0])[:5] #Mencetak data yang telah difilter
```
