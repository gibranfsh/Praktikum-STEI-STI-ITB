```python
import pandas as pd

df = pd.read_excel("WorldCups.xlsx")
```

```python
df1 = df.loc[df["QualifiedTeams"] > 20]
df1.plot(kind="line",x="Year",y="Attendance")
```

```python
df1["Winner"].value_counts().plot(kind = "pie", ylabel = " ", autopct = "%1.0f%%") #Mencetak pie chart
```

```python

df2 = df.groupby(["Organization"]).count() #Menggunakan parameter groupby utk jumlah organisasi sub fifa lalu menyimpannya dalam suatu variabel
df2["Year"].plot(kind = "barh") #Mencetak Barh karena lebih mudah untuk melihat berapa kali sub organisasi fifa tersebut menjadi penyelenggara
```
