```python
#Subsoal 1

df[["nilai_mat"]].plot(kind = "hist", bins = [0 ,20 ,40 ,60 ,80 ,100], title = "Distribusi Nilai Matematika", rwidth = 0.8, color = "green") #Histogram distribusi nilai matematika
```

```python
#Subsoal 2

banyakmhsw = df.groupby(["fakultas"]).count() #Menggunakan parameter groupby utk jumlah fakultas lalu menyimpannya dalam suatu variabel
banyakmhsw["nama"].plot(kind = "barh", title = "Banyak Mahasiswa Masing-masing Fakultas") #Mencetak horizontal bar dengan sumbu-x sebagai fakultas dan sumbu-y sebagai jumlah mahasiswa

```

```python
#Subsoal 3

df["fakultas"].value_counts().plot(kind = "pie", title = "Banyak Mahasiswa Masing-masing Fakultas")
```

```python
#Subsoal 4

#Berdasarkan diagram pie dan batang horizontal, fakultas dengan mahasiswa terbanyak adalah FTI dengan batang horizontal sebagai
#diagram yang lebih baik dalam menampilkan fakultas dengan mahasiswa terbanyak disebabkan dalam batang horizontal terdapat
#jumlah mahasiswa yang lebih terlihat/tampak yang ditandai dengan rentang angka, sedangkan pada diagram pie
#tidak tercantum jumlah mahasiswanya, hanya ada bentuk pie yang hanya dapat dilihat bentuk besar atau kecilnya saja.
```
