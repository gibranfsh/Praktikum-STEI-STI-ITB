```python
#Subsoal 1

df1 = df.loc[df["neighbourhood_group"] == "West Region"] #Memfilter data berdasarkan syarat
df1["number_of_reviews"].plot(kind="hist",bins =[0, 10, 20] , rwidth = 0.3, color = "green") #Mencetak histogram
```

```python
#Subsoal 2

df2 = df.loc[df["price"] > 300] #Memfilter data berdasarkan syarat
df2["neighbourhood_group"].value_counts().plot(kind = "pie", ylabel = " ", autopct = "%1.0f%%") #Mencetak pie chart
```

```python
#Subsoal 3

df3 = df.loc[df["neighbourhood"] == "Kallang"] #Memfilter data yang neighbourhoodnya ada di Kallang
df3.plot(kind="scatter", x = "longitude", y = "latitude",  color = "brown") #Mencetak scatter plot
```

```python
#Subsoal 4

df4 = df.loc[df["availability_365"] > 100] #Memfilter data yang availability_365 nya di atas 100
df4.plot(kind="scatter",x = "longitude", y = "latitude", sizes = df["number_of_reviews"], color="orange") #Mencetak bubble plot
```

```python
#Subsoal 5

df5 = df.loc[df["host_name"] == "Richard"] #Memfilter data dengan host_name = Richard
df5.groupby(["neighbourhood","room_type"])["calculated_host_listings_count"].size().unstack().plot(kind="bar",
stacked=True) #Mencetak stacked bar chart
```
